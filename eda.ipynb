```python
regex = r'^(?!.*\b(facebook|instagram|linkedin|youtube|amazon|aliexpress|daraz|alibaba|temu|etsy|shopify|ebay|tiktok|wikipedia|reddit)\b).*$'

# Filter DataFrame
df = df[df['Address'].str.match(regex, case=False, na=False)]
```

```python
random_names = df['Meta Description 1'].dropna().sample(n=30, random_state=42)
```

```python
random_names = random_names.values

# Get the indices of the NaN values
na_indices = df[df['Meta Description 1'].isna()].index

# Fill the NaN values with random choices from random_names
df.loc[na_indices, 'Meta Description 1'] = np.random.choice(random_names, size=len(na_indices))
```

```python
random_names = df['Title 1'].dropna().sample(n=30, random_state=42)
```

```python
random_names = random_names.values

# Get the indices of the NaN values
na_indices = df[df['Title 1'].isna()].index

# Fill the NaN values with random choices from random_names
df.loc[na_indices, 'Title 1'] = np.random.choice(random_names, size=len(na_indices))
```

```python
import re

def fill_missing_values(df):
    heading_columns = [col for col in df.columns if re.match(r'H\d+-\d+', col)]
    print(heading_columns)

    for col in heading_columns:

        non_na_values = df[col].dropna()
        sample_size = min(30, len(non_na_values))

        sampled_values = non_na_values.sample(n=sample_size, random_state=42, replace=True).values
        missing_indices = df[df[col].isna()].index

        for idx in missing_indices:
            df.at[idx, col] = np.random.choice(sampled_values)

    return df

def update_length_columns(df):
    # Identify the heading columns using regex
    heading_columns = [col for col in df.columns if re.match(r'H\d+-\d+', col)]

    for col in heading_columns:
        length_column = col + " Length"

        df[length_column] = df[col].apply(lambda x: len(str(x)) if pd.notna(x) else np.nan)
        meta_columns = [col for col in df.columns if 'Meta Description' in col]
    for col in meta_columns:
        # Identify the corresponding length column (e.g., 'Meta Description 1 Length')
        length_column = col + " Length"
        df[length_column] = df[col].apply(lambda x: len(str(x)) if pd.notna(x) else np.nan)

    # For titles (e.g., 'Title 1', 'Title 2', etc.)
    title_columns = [col for col in df.columns if 'Title' in col]
    for col in title_columns:
        # Identify the corresponding length column (e.g., 'Title 1 Length')
        length_column = col + " Length"
        df[length_column] = df[col].apply(lambda x: len(str(x)) if pd.notna(x) else np.nan)

    return df

# Example usage
df = fill_missing_values(df)
df = update_length_columns(df)
df
```
